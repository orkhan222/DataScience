```python
# import
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel(r'../Datasets/test_dataset_new.xlsx')
df.head(1)
```

```python
# Delete if price= 'Not found'
df['prc-dsc'] = (df['prc-dsc'].str.replace(' TL', '').str.replace('.', '').str.replace(',', '.').astype(float, errors='ignore'))
print(df)
```

```python
df = df[df['prc-dsc'].notna()]
```

```python
df = df.replace('Not found', pd.NA)
print(df)
```

```python
threshold = 0.5 * len(df)
df = df.dropna(thresh=threshold, axis=1)
print(df)
```

```python
df['prc-dsc'] = pd.to_numeric(df['prc-dsc'], errors='coerce')
print(df)
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    if len(df[col].unique()) <= 10:
        df = pd.get_dummies(df, columns=[col], prefix=[col], drop_first=True)
    else:
        lbl = LabelEncoder()
        df[col] = lbl.fit_transform(df[col].astype(str))
df.head()
```

```python
correlation = df.corr()['prc-dsc'].abs()
low_correlation_cols = correlation[correlation < 0.20].index
data = df.drop(columns=low_correlation_cols)
print(data)
```

```python
def delete_rows_with_not_found(df, column_name, price_to_delete):
    new_price = df[column_name] != price_to_delete
    df = df[new_price]
    
    return df
price_to_delete = 'Not found'

df = delete_rows_with_not_found(df, 'price', price_to_delete)

print(df)
```

```python
# Function to split price column into currency and amount
def split_price_column(df):
    df['amount'] = df['price'].str.replace('.', '', 1)
    df['amount'] = df['price'].str.replace(',', '.', regex=False)
    df[['amount', 'currency']] = df['price'].str.split(' ', expand=True)
    return df
df = split_price_column(df)

print(df)
```
