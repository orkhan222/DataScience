```python
# import
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv(r'../Datasets/OnlineRetail.csv',encoding='ISO-8859-1')
df.head(5)
```

```python
# Second Task: Take a raw dataset and perform feature engineering. 
# This can include creating new features, transforming existing features, 
# and encoding categorical variables.
# Explain the rationale behind each decision.
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['DayOfWeek'] = df['InvoiceDate'].dt.dayofweek
print(df)
```

```python
# Second Task: Take a raw dataset and perform feature engineering. 
# This can include creating new features, transforming existing features, 
# and encoding categorical variables.
# Explain the rationale behind each decision.
data = df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
print(data)
```

```python
# Second Task: Take a raw dataset and perform feature engineering. 
# This can include creating new features, transforming existing features, 
# and encoding categorical variables.
# Explain the rationale behind each decision.
customer_features = df.groupby('CustomerID').agg(
                   TotalSpend=('TotalPrice', 'sum'),
                 AverageSpend=('TotalPrice', 'mean'),
                   TotalItems=('Quantity', 'sum')
).reset_index()
data = df.merge(customer_features, on='CustomerID', how='left')
print(data)
```

```python
# Second Task: Take a raw dataset and perform feature engineering. 
# This can include creating new features, transforming existing features, 
# and encoding categorical variables.
# Explain the rationale behind each decision.
product_features = df.groupby('StockCode').agg(
    TotalQuantitySold=('Quantity', 'sum'),
    AveragePrice=('UnitPrice', 'mean')
).reset_index()
data = df.merge(product_features, on='StockCode', how='left')
print(data)
```
