```python
# import
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
```

```python
df = pd.read_csv(r'../Datasets/w5_application_data_small.csv')
df.head(5)
```

```python
numerical = data.select_dtypes(include=['int64', 'float64']).columns
categoricals = data.select_dtypes(include=['object']).columns
print(numerical)
print(categoricals)
```

```python
anova_results = {}
```

```python
for column in numerical:
    if column != 'TARGET':  
        group1 = data[data['TARGET'] == 0][column].dropna()
        group2 = data[data['TARGET'] == 1][column].dropna()
        f_stat, p_value = f_oneway(group1, group2)
        anova_results[column] = {'F-statistic': f_stat, 'p-value': p_value}
```

```python
results = pd.DataFrame(anova_results).T
results.sort_values(by='F-statistic', ascending=False, inplace=True)
print(results)
```
