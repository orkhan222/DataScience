```python
# import
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'../Datasets/shopping_trends.csv')
df.head(5)
```

```python
# Bar Charts
category = df['Category'].value_counts()
plt.figure(figsize=(10, 6))
category_bar = sns.barplot(x=category.index, y=category.values, palette='viridis')
plt.title('Number of Items Purchased in Each Category', fontsize=15)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of Items Purchased', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#  Line Charts
age_purchase = df.groupby('Age')['Purchase Amount (USD)'].mean()
plt.figure(figsize=(12, 6))
age_line = plt.plot(age_purchase.index, age_purchase.values, marker='o', color='blue', linestyle='-', linewidth=2, markersize=5)
plt.title('Average Purchase Amount (USD) by Age', fontsize=15)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Average Purchase Amount (USD)', fontsize=15)
plt.grid(True)
plt.show()
```

```python
# Scatter Plots
plt.figure(figsize=(10, 6))
rating_scatter = sns.scatterplot(data=df, x='Review Rating', y='Purchase Amount (USD)', hue='Category', palette='deep', s=50)
plt.title('Relationship Between Review Rating and Purchase Amount', fontsize=15)
plt.xlabel('Review Rating', fontsize=12)
plt.ylabel('Purchase Amount (USD)', fontsize=12)
plt.show()
```
