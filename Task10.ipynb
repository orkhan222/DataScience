```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv("/content/public_transportation_data.csv")

# Assuming you have a column named "Country" in your DataFrame
country_counts = data['origin'].value_counts()

# Create a bar chart to visualize the visit counts
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Train Visits by Country')
plt.xlabel('origin')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the CSV file
data = pd.read_csv("/content/public_transportation_data.csv")

# Assuming you have columns "Country" and "Train Type" in your DataFrame
# Group the data by both Country and Train Type and count the occurrences
grouped_data = data.groupby(['origin', 'train type']).size().unstack(fill_value=0)

# Create a grouped bar chart to visualize the train visits by country and train type
plt.figure(figsize=(12, 6))
sns.heatmap(grouped_data, cmap='Blues', annot=True, fmt='d')
plt.title('Train Visits by Country and Train Type')
plt.xlabel('Train Type')
plt.ylabel('Country')

# Display the chart
plt.show()
```
