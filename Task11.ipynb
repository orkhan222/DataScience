```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
import pandas as pd
data = pd.read_csv("public_transportation_data.csv")
data
```

```python
country_counts = data['destination'].value_counts()

# Display the countries and their visit counts
print(country_counts)
```

```python
date_counts = data['insert_date'].value_counts()

# Display the date
print(date_counts)
```

```python
country_counts = data['origin'].value_counts()

# Create a bar chart to visualize the visit counts
plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Train Visits by Country')
plt.xlabel('origin')
plt.ylabel('Number of Visits')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the chart
plt.show()
```

```python
train_class_price_mean = data.groupby('train_class')['price'].mean()

# Display the average price for each train class
print(train_class_price_mean)
```

```python
df = pd.read_csv('/content/public_transportation_data.csv')

df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])

# Calculate the duration for each row
df['duration'] = (df['end_date'] - df['start_date']).dt.total_seconds() / 3600  # Duration in hours

# Find the destination with the longest duration
longest_destination = df.loc[df['duration'].idxmax()]

print(f"The destination that took the longest time is '{longest_destination['destination']}' with a duration of {longest_destination['duration']:.2f}Â hours.")
```

```python
# Create a count plot to visualize the popularity of train types
plt.figure(figsize=(15, 6))  # Adjust the figure size as needed
sns.countplot(data=data, x='train_type')
plt.title('Popularity of Train Types')
plt.xlabel('Train Type')
plt.ylabel('Count')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Display the chart
plt.show()
```
